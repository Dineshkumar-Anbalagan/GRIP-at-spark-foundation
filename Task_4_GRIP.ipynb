### **IMPORTING LIBRARIES**

```
#IMPORTING LIBRARIES:
#Importing data manipulation and analysis library:
import pandas as pd

#Importing scientific computing library:
import numpy as np

#Importing Visualization libraries:
import matplotlib.pyplot as plt
import seaborn as sns
print('Libraries Imported')
```

### **LOADING DATASET**

```
#Importing Dataset:
Data=pd.read_csv('SampleSuperstore.csv')
Data.head()
```

### **ANALYSING DATASET**

```
#ANALYSING THE DATASET:
print('Dimension of dataset:{}'.
      format(Data.shape),'\n',75*'-')
print('name of the columns:{}'.
      format(Data.columns),'\n',75*'-')
print('Number of duplicated values:{}'.
      format(Data.duplicated().sum()),'\n',75*'-')
print('rows contain null values:\n{}'.
      format(Data.isnull().sum()),'\n',75*'-')
print('Schema of the dataset:\n{}'.
      format(Data.info()),'\n',75*'-')
print('Descriptive Analysis:\n{}'.
      format(Data.describe()),'\n',75*'-')
print('Correlation:\n{}'.
      format(Data.corr()))
```

```
#Removing duplicate values:
Data=Data.drop_duplicates()
```

### **EDA**

```
#correlation matrix:
plt.figure(figsize=(8,5))
correlation=Data.corr()
sns.heatmap(correlation,annot=True,cmap='Purples')
```

```
plt.figure()
sns.pairplot(Data)
```

```
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=Data)
plt.legend('Profit')
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(30,13))
plt.plot(Data['Profit'], color='blue')
plt.grid(alpha=0.3)
```

```
%matplotlib inline
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.countplot('Segment',data=Data)
plt.title('DISTRIBUTION OF SEGMENTS')
plt.xlabel('SEGMENT',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)
plt.subplot(1,2,2)
sns.countplot('Ship Mode',data=Data)
plt.title('DISTRIBUTION OF SHIP MODE')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('SHIP MODE',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

### **OBSERVATION:**
####  **1. From the above "DISTRIBUTION OF SEGMENTS" plot it is clear that the "*consumers*" are the highest buyers of goods and "*Home office*" are the least buyers of goods**
#### **2. From the above "DISTRIBUTION OF SHIP MODE" plot, it is clear that the "standard class" mode is the highest mode for shipping goods and "same day" mode is the least mode for shipping goods**

### **DISTRIBUTION OF SHIPMENTS BASED ON STATES**

```
plt.figure(figsize=(30,10))
sns.countplot(Data['State'])
plt.xticks(fontsize=16,rotation=90)
plt.ylabel('COUNTS',fontsize=20)
plt.xlabel('STATES',fontsize=20)
plt.title('DISTRIBUTION OF STATES',fontsize=25);
plt.grid(alpha=0.3)
```

### **OBSERVATION:** 
#### **From observing the above distribution plot, it is clear that the goods shipped to California is higher than any other states in the United States, in other words California has the highest rate of buyers comparing to other states and West Virginia has the least rate of buyers in the entire United States**

### **DISTRIBUTION BASED ON REGION/CATEGORY**

```
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.countplot('Region',data=Data)
plt.title('DISTRIBUTION OF REGIONS')
plt.xlabel('REGION',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)
plt.subplot(1,2,2)
sns.countplot('Region',hue='Category',data=Data)
plt.title('DISTRIBUTION OF REGIONS BASED ON CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

### **OBSERVATION:**
#### **From the above plots it is clear that the Western states has the highest rate of buyers while southern states has the lowest rate of buyers. From our previous observations we found out that the California has the highest rate of buyers - in this case California also a part of western states and this proves our statement.**

### **DISTRIBUTION BASED ON CATEGORIES/SUB-CATEGORIES** 

```
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
sns.countplot('Category',data=Data);
plt.title('DISTRIBUTION OF CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
plt.subplot(2,1,2)
sns.countplot('Sub-Category',data=Data);
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
```

```
plt.figure(figsize=(18,10))
sns.countplot('Category',hue='Sub-Category',data=Data);
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
```

### **OBSERVATION:** 
#### **From the above plots it is clear that the office supplies has the highest rate of shippings in total. Here, Binders has the highest shipping count while copiers has the least shipping count under office supplies.**

### **PROFIT DISTRIBUTION AMONG STATES**

```
plt.figure(figsize=(30,10))
sns.barplot(x=Data['State'].sort_values(ascending=True),y=Data['Profit'])
plt.title('DISTRIBUTION OF PROFITS BASED ON STATES', fontsize=20)
plt.xticks(rotation=90, fontsize=15);
plt.xlabel('STATE',fontsize=15)
plt.ylabel('PROFIT',fontsize=15)
plt.grid(alpha=0.3)
```

### **PROFIT DISTRIBUTION BASED ON DIFFERNT REGIONS**

```
C=Data[Data['Region']=='Central']
E=Data[Data['Region']=='East']
W=Data[Data['Region']=='West']
S=Data[Data['Region']=='South']
```
